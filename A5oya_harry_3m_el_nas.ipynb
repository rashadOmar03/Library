```python
class Library:
    def __init__(self,name_of_library,booklist):
        self.booklist=booklist
        self.name_of_library=name_of_library
        self.lended_books={}


    def display(self):
        return self.booklist


    def add_books(self,books):
        if type(books) == str and books not in self.booklist:
            self.booklist.append(books)
            return self.booklist
        else:
            print("please enter a valid type string")


    def lend(self, name, b_name):
        if type(name) == str:
            b_name_lower = b_name.lower()
            for b in self.booklist:
                if b.lower() == b_name_lower:
                    self.booklist.remove(b)
                    if name not in self.lended_books:
                        self.lended_books[name] = [b]
                    else:
                        self.lended_books[name].append(b)
                    return self.lended_books
            
            print(f"book is not in the booklist:{self.booklist}")
        else:
            print("please enter string:")


    def ret_borrowed(self, n, b):
        b_lower = b.lower()
        if n in self.lended_books:
            borrowed_books = self.lended_books[n]
            for i in borrowed_books:
                if i.lower() == b_lower:
                    self.lended_books[n].remove(i)
                    self.booklist.append(i)
                    print("Success! Book returned.")
                    return self.booklist
            else:
                print(f"{n} did not borrow {b} book.")
        else:
            print(f"{n} did not borrow any books.")
```

```python
harry = Library("CodeWithHarry", ['Python', 'Rich Daddy Poor Daddy', 'Harry Potter', 'C++ Basics', 'Algorithms by CLRS'])
```

```python
print("Welcome to CodeWithHarry Library!")
while True:
    print("choose a number:")
    print("1. Display Books")
    print("2. Lend a Book")
    print("3. Add a Book")
    print("4. Return a Book")
    print("5. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        print(harry.display())
    elif choice == "2":
        book = input("Enter the book you want to lend: ")
        user = input("Enter your name: ")
        harry.lend(user,book)
    elif choice == "3":
        book = input("Enter the book you want to add: ")
        harry.add_books(book)
    elif choice == "4":
        book = input("Enter the book you want to return: ")
        user = input("Enter your name: ")
        harry.ret_borrowed(user,book)
    elif choice == "5":
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please enter only digits.")
    
```
